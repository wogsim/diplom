```python
top_channels = df['channel_handle'].value_counts().index
df_top = df[df['channel_handle'].isin(top_channels)]

g = sns.FacetGrid(df_top, col="channel_handle", col_wrap=3, sharex=True, sharey=False)
g.map(sns.histplot, "reward", bins=30, kde=True)
g.set_axis_labels("Z-score (Награда)", "Количество постов")
plt.show()
```
